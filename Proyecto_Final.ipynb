```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot as plt
import plotly.express as px

# Leer los datos desde los archivos
datos1 = pd.read_csv('Encuesta-Salud.csv')
datos2 = pd.read_csv('Predios-Titulados.csv')
datos3 = pd.read_csv('MEN_ESTADISTICAS_EN_EDUCACION.csv')
datos4= pd.read_csv('Saber_11_2019.csv')
```

```python
datos1
```

```python
# Contar la cantidad de personas por género
conteo_genero = datos1['2. Género'].value_counts()

# gráfico de barras
plt.figure(figsize=(10, 6))
ax = conteo_genero.plot(kind='bar', color=['lightblue', 'pink'])  # 'blue' para hombres, 'pink' para mujeres
plt.title('Cantidad de personas por género en encuesta de salud')
plt.xlabel('Género')
plt.ylabel('Cantidad de personas')
plt.xticks(rotation=0)

# Agregar etiquetas a las barras
for i, v in enumerate(conteo_genero):
    plt.text(i, v + 0.1, str(v), ha='center', va='bottom')

plt.show()
```

```python
conteo_genero = datos1['5. ¿Cuántos casos conoces de embarazos en adolescentes?'].value_counts()

# gráfico de barras
plt.figure(figsize=(10, 6))
conteo_genero.plot(kind='bar')
plt.title('Percepción de embarazos adolescentes')
plt.xlabel('Respuestas')
plt.ylabel('Cantidad de respuestas')
plt.xticks(rotation=0)

# Agregar etiquetas a las barras
for i, v in enumerate(conteo_genero):
    plt.text(i, v + 0.1, str(v), ha='center', va='bottom')
plt.show()
```

## Conclusiones:
•	En el gráfico se puede observar que la mayoría de los encuestados no tiene información sobre embarazos adolescentes en su comunidad.

•	Aproximadamente el 22% cree que solo uno de cada diez adolescentes ha experimentado un embarazo.

•	Un 30% de los encuestados cree que más de dos de sus compañeros adolescentes han experimentado un embarazo.


```python
conteo_genero = datos1['22. ¿Conoces niñas entre 12 y 18 años de edad que han sido obligadas a tener relaciones sexuales?'].value_counts()

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
conteo_genero.plot(kind='bar', color=['lightblue', 'pink'])  # 'blue' para hombres, 'pink' para mujeres
plt.title('¿Conoces niñas entre 12 y 18 años de edad que han sido obligadas a tener relaciones sexuales?')
plt.xlabel('Respuestas')
plt.ylabel('Cantidad de respuestas')
plt.xticks(rotation=0)
plt.show()
```

## Datos estadísticas en educación en Colombia

```python
datos3
```

Filtramos para ver estadísticas especificamente en Bogotá.

```python
datos_bogota = datos3[datos3['DEPARTAMENTO'] == 'Bogotá, D.C.']
```

```python
plt.figure(figsize=(12, 6))
plt.plot(datos_bogota['AÑO'], datos_bogota['TASA_MATRICULACIÓN_5_16'])
plt.title('Tasa de Matriculación en Bogotá a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Tasa de Matriculación')
plt.show()
```

Luego quisimos ver las mismas estadísticas pero esta vez para medellín.

```python
datos_medellin = datos3[datos3['DEPARTAMENTO'] == 'antioquia']

plt.figure(figsize=(12, 6))
plt.plot(datos_bogota['AÑO'], datos_bogota['TASA_MATRICULACIÓN_5_16'], label='Bogotá')
plt.plot(datos_medellin['AÑO'], datos_medellin['TASA_MATRICULACIÓN_5_16'], label='Antioquia')
plt.title('Tasa de Matriculación en Bogotá y Antioquia a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Tasa de Matriculación')
plt.legend()
plt.show()
```

## Conclusiones
•	En el gráfico se puede observar que Bogotá para el año 2019 tuvo una tasa de matrícula consistentemente mayor que la de Antioquia.

•	De igual manera también se puede observar que, antes del 2018, Bogotá tenía tasas de matrícula muy bajas respecto a Antioquia y que de alguna manera esto mejoró a partir de ese año, sería importante entrar a revisar, que causó ese aumento.

•	Para el año 2022 se ve que ambos lugares, presentan una tasa de matriculación muy cercana.


# Correlación entre la tasa de matriculación y cobertura neta en Bogotá

```python
plt.figure(figsize=(9, 6))
sns.regplot(x='TASA_MATRICULACIÓN_5_16', y='COBERTURA_NETA', data=datos_bogota)
plt.title('Correlación entre La Tasa de Matriculación y la cobertura neta en Bogotá')
plt.xlabel('Tasa de Matriculación')
plt.ylabel('Cobertura Neta')
plt.show()
```

# Conclusiones:
•	Existe una correlación positiva entre la tasa de matrícula y la cobertura neta en Bogotá. Esto significa que a medida que aumenta la cobertura neta también aumenta la tasa de matrícula. Esto indica que existe una relación estrecha entre las dos variables.

•	La tendencia es lineal, lo que significa que la relación entre las dos variables es constante.


```python
fig = px.scatter(datos_bogota, x='TASA_MATRICULACIÓN_5_16', y='COBERTURA_NETA', title='Gráfico interactivo de la tasa de matriculación vs. la cobertura neta en Bogotá')

fig.update_layout(
    width=900,  # Ancho en píxeles
    height=600,  # Alto en píxeles
)


fig.show()
```

```python
datos_bogota = datos3[datos3['DEPARTAMENTO'] == 'Bogotá, D.C.']
```

```python
plt.figure(figsize=(12, 6))
plt.plot(datos_bogota['AÑO'], datos_bogota['TASA_MATRICULACIÓN_5_16'])
plt.title('Tasa de Matriculación en Bogotá a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Tasa de Matriculación')
plt.show()
```

```python
datos_medellin = datos3[datos3['DEPARTAMENTO'] == 'antioquia']

plt.figure(figsize=(12, 6))
plt.plot(datos_bogota['AÑO'], datos_bogota['TASA_MATRICULACIÓN_5_16'], label='Bogotá')
plt.plot(datos_medellin['AÑO'], datos_medellin['TASA_MATRICULACIÓN_5_16'], label='Antioquia')
plt.title('Tasa de Matriculación en Bogotá y Antioquia a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Tasa de Matriculación')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9, 6))
sns.regplot(x='TASA_MATRICULACIÓN_5_16', y='COBERTURA_NETA', data=datos_bogota)
plt.title('Correlación entre La Tasa de Matriculación y la cobertura neta en Bogotá')
plt.xlabel('Tasa de Matriculación')
plt.ylabel('Cobertura Neta')
plt.show()
```

```python
fig = px.scatter(datos_bogota, x='TASA_MATRICULACIÓN_5_16', y='COBERTURA_NETA', title='Gráfico interactivo de la tasa de matriculación vs. la cobertura neta en Bogotá')

fig.update_layout(
    width=900,  # Ancho en píxeles
    height=600,  # Alto en píxeles
)


fig.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='TASA_MATRICULACIÓN_5_16', y='COBERTURA_NETA', data=datos3)
plt.title('Correlación entre La Tasa de Matriculación y la cobertura neta en todo el territorio nacional')
plt.xlabel('Tasa de Matriculación')
plt.ylabel('Cobertura Neta')
plt.show()
```

```python


# Crear una figura
plt.figure(figsize=(15, 6))

# Obtener la lista de ciudades únicas
ciudades = datos3['DEPARTAMENTO'].unique()

# Iterar sobre cada ciudad y trazar la línea correspondiente
for ciudad in ciudades:
    # Filtrar los datos para la ciudad actual
    datos_ciudad = datos3[datos3['DEPARTAMENTO'] == ciudad]
    # Trazar la línea para la ciudad actual
    plt.plot(datos_ciudad['AÑO'], datos_ciudad['TASA_MATRICULACIÓN_5_16'], label=ciudad)

# Establecer el título y las etiquetas de los ejes
plt.title('Tasa de Inflación a lo largo de los años por Ciudad')
plt.xlabel('Año')
plt.ylabel('Tasa de Matriculación (5-16 años)')

# Mostrar la leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```
